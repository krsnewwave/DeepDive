# Breadth of NLP
## NLP Systems
NLP can be thought of as a spectrum, from rule-based systems to probablistic systems.

1) Rule-based systems
   - historical NLP systems took in a lot of linguistic literature, particularly grammars
   
2) Probablistic systems
   - modern NLP takes into account massive text or audio datasets, commonly called corpora in literature. Words, characters, thought and topics are taken as events, and probabilities are computed according to empirical observations in the corpus.
   
3) Hybrid systems
   - the more successful systems are not wholly rule-based nor probablistic. Grammars are inherent in human conversation, and advances in linguistics are complimented by computation.

# How about some supervised learning?
This is an example of data whose features is unknown. Obviously, we use the words, not the characters, since a single word expresses <b>a thought that can explain the category</b>. On the other hand, these English or Filipino characters don't. However, we have another challenge. How do we represent these words as features? This is the problem of feature representation, and here we introduce the bag-of-words (BOW) representation.

BOW is a matrix defined by $\mathbb{M} \in \mathbb{R}^{mxn}$ where m represents documents in the corpus for $i=0...m$ and n represent word features for $j=0...n$. Matrix values $c_{i,j}$ could be ${0,1}$, $\mathbb{Z}$ (counts) or $\mathbb{Q}^+$ (frequencies).

In BOW, each word is then treated separately. There are pros and cons in such a representation:

Pros:

1) Very fast to compute.

2) Easy to store.

Cons:

1) Words are not totally independent. N-grams can serve to alleviate this, but what about long term dependencies? Consider the following: 

    Jack, although poor, received honors for their academic performance.

2) Word-sense disambiguation. Consider how monstrous is used in following sentences:

    You shuddered as you gazed, and wondered what monstrous cannibal and savage could ever have gone a death-harvesting with such a hacking, horrifying implement.
    
    She is a monstrous lucky girl to get him, upon my honour; not but that he is much more lucky in getting her, because she is so very handsome and agreeable, that nothing can be good enough for her.

# New Model: Naive Bayes
The Naive Bayes classifier is a probablistic classifier, using the maximum likelihood estimates and the Bayes rule.
(reference: cs.colostate.edu, nlp.stanford.edu)

## What is maximum likelihood?

Given feature vector $x^{(1)} = (x_1, x_2, ... x_n)$:

  $$ 
  \begin{align}
  P(x^{(1)}|\theta) & = P(x_1 | \theta_1) P(x_2 | \theta_2) ... P(x_n | \theta_n) \\
  & = \prod_{i=1}^{n}P(x_i|\theta_i)
  \end{align}
  $$
  
  We say this is a maximum likelihood solution since we find $\theta$ that maximize $P(X|\theta)$.

## Classification

  Now, applying this to classification, we have in probability:
  
  $$
P(Y|X) = \frac{P(X|Y) P(Y)}{P(X)}
$$

where $P(Y|X)$ is called the posterior, $P(X|Y)$ is called the likelihood and $P(Y)$ is called the prior.

For classification, $P(X)$ is not so important since we'll be only using the <b>maximum a posteriori (MAP)</b> rule, that is:

$$
\begin{align}
\operatorname{arg\,max}_y P(Y|X) & = \operatorname{arg\,max}_y \frac{P(X|Y) P(Y)}{P(X)}\\
& = \operatorname{arg\,max}_y P(X|Y) P(Y)
\end{align}
$$

Taking this into text classification:

$$
P(c|d) \propto P(c) \prod_{1 \le k \le n_d} P(t_k|c) 
$$

Since this can cause underflow, we use sum of log probabilities:

$$
P(c|d) \propto log P(c) + \sum_{1 \le k \le n_d} log P(t_k|c) 
$$

where $P(t_k|c)$ is the conditional probability of term $t_k$ appearing in a document of class c. That is,

$$
\hat{P}(t_k|c) = \frac{T_{ct}}{\sum_{t \in V} T_{ct}}
$$

where $T_{ct}$ is the number of occurences of term i in training documents from class c. We're discarding where term t appeared in the document.

Lastly, we just get the MAP class:
$$
c_{map} = \operatorname{arg\,max}_{c \in \mathbb{C}} log P(c) + \sum_{1 \le k \le n_d} log P(t_k|c) 
$$

### Laplace Smoothing
What if a word in the vocabulary does not appear in the dictionary? Thus $\hat{P}(t_k|c)$ becomes 0. We generally don't want zeroes in our computations. Thus, we do smoothing: 

$$
\hat{P}(t_k|c) = \frac{T_{ct}+1}{\sum_{t \in V} T_{ct}+|V|}
$$

### Pros and Cons:

<b>Pros:</b>

1) NB is probably the easiest classifiers to implement and understand.

2) Storage-wise, it's very effective.

3) Computation at testing time is also very fast.

<b>Cons:</b>

1) Extending it may require understanding different statistical distributions (more of a caveat really)

2) Requires I.I.D, that is, all features are independent of each other. 

Even though NB is simple, it has worked well in practice, particularly, text classification, where there are a lot of words. It does not suffer as much from the curse of dimensionality since the decoupling of the class conditional feature distributions means that each distribution can be independently estimated as a one dimensional distribution. # TF-IDF Weighting
TF-IDF is used to strike a balance between rare words and common words. What we want really is to find words that are <b>important</b> to a class, even if it is rare. It is computed by the following:

$$
\text{tf-idf}_{t,d} = \text{tf}_{t,d} * idf_t\\
$$

$$
idf_t = log \frac{N}{df_t}
$$

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer = TfidfVectorizer(stop_words=all_stopwords)
vectorizer
```

```python
y_tfidf = [sens.categories(fileids=f)[0] for f in files]

files = [f for f in sens.fileids()]
corpus = [sens.raw(fileids=[f]) for f in files]

X_tfidf = vectorizer.fit_transform(corpus)
X_tfidf.shape
```

```python
X_tfidf_train, X_tfidf_test, \
    y_tfidf_train, y_tfidf_test = cv.train_test_split(X_tfidf, y_tfidf, 
                                          test_size=0.33)
```

```python
scores = cv.cross_val_score(MultinomialNB(), X_tfidf,y_tfidf,cv=10)
print scores
print np.mean(scores)
print np.std(scores)
```

```python
# train the classifier
classifier = MultinomialNB()
classifier.fit(X_tfidf_train, y_tfidf_train)

# predict labels for the test set
predictions = classifier.predict(X_tfidf_test)

# output the classification report
label_names = list(set(y_tfidf_train))
print(classification_report(y_tfidf_test, predictions, target_names=label_names))
```
