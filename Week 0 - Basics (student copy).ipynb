##  Generators and list comprehensions

Generators and list comprehensions are pythonic structures that can aid both in readability and efficiency. Recall that Python is as much as a functional language as it is an object-oriented language. There's so much good as well to be found in functional languages.

Don't worry if you don't understand too much yet. We'll be encountering this frequently until we get the hang of it.

```python
def gen_num_up_to(n):
    num = 0
    while num < n:
        yield num
        num +=1        
```

```python
[n for n in gen_num_up_to(10)]
```

## Slicing

Slicing is an interesting way to index iterables. Again, it's a surprising syntax, but we'll be using this until we get the hang of it.

```python
x = [n for n in gen_num_up_to(10)]
```

```python
print(x)
print(x[:5])
print(x[5:])
print(x[3:8])
print(x[:-2])
```

# I. Need. Pictures.

Matplotlib, specifically pyplot, is a collection of functions that can let us visualize our data right here in the notebook.

```python
from matplotlib import pyplot as plt
%matplotlib inline
```

Let's study the histogram of the distribution of classes in the dataset. Recall that we can extract the column as a list. Then, we'll get the frequencies of these classes. Afterwards, we submit these 2 as input to the histogram.

Note, we're using a list comprehension to get the frequencies from the Counter. We're saying the following: <b> For each element in the dictionary (counter), get the second element, which is the frequency </b>.

```python
from collections import Counter
c = Counter(data["Name"])
cats = list(np.unique(data["Name"]))
freqs = [v for k,v in c.iteritems()]
freqs
```

```python
plt.bar(range(3), freqs)
plt.xticks(range(3), cats)
plt.show()
```

Next, we'll plot the petal lengths. This variable is a continuous variable. it is different from the above where we plot nominal or discrete variables. PyPlot has a specific function that automatically creates the bins for us. Of course, we can make the binning ourselves if we need to.

```python
plt.hist(data.PetalLength)
plt.show()
```

Now that isn't too interesting isn't it?  All we know is that there are two peaks. Let's make it more interesting. We'll plot the petal lengths of each flower next. We'll get the individual lengths of each class, then plot them one by one into a kind of stacked chart.

```python
ax = plt.subplot(111)

p_length_setosa = data[data.Name=="Iris-setosa"]["PetalLength"]
p_length_virginica = data[data.Name=="Iris-virginica"]["PetalLength"]
p_length_versicolor = data[data.Name=="Iris-versicolor"]["PetalLength"]

ax.hist(p_length_setosa, width=0.2,color='b', label="Setosa")
ax.hist(p_length_virginica, width=0.2,color='g', label="Virginica")
ax.hist(p_length_versicolor ,width=0.2,color='r', label="Versicolor")

plt.title("Petal Lengths per class")
plt.legend()
plt.show()
```

There it is. It seems classifying setosa is very easy given this histogram. It's the virginica and versicolor that needs some work. By the way, we can make another very interesting visualization using pandas. It's a higher level interface to PyPlot.

```python
from pandas.tools.plotting import scatter_matrix
color_wheel = {"Iris-setosa": "blue", "Iris-virginica":"red", "Iris-versicolor":"green"}
scatter_matrix(data, c=data["Name"].apply(lambda x:color_wheel[x]), alpha=0.2, figsize=(20, 20), 
               diagonal='kde', marker='o')
plt.show()
```

It seems that Iris setosa is the easiest to classify. Virginica and versicolor though is a harder story.

#  Feature Engineering Board games

Feature engineering is a task in data science that underscores most of the field. It is concerned with <b> getting the variables extracted and represented appropriately </b> for further tasks in the pipeline to be carried out. There is perhaps nothing more fundamental than feature engineering (save perhaps getting your data cleaned up!) in the whole of data science.

Please download the following https://github.com/ThaWeatherman/scrapers/blob/master/boardgamegeek/games.csv. This dataset is from boardgamegeek, which is an online site for board game enthusiasts. The columns are defined like so:

<ul>
    <li>name – name of the board game.</li>
    <li>playingtime – the playing time (given by the manufacturer).</li>
    <li>minplaytime – the minimum playing time (given by the manufacturer).</li>
    <li>maxplaytime – the maximum playing time (given by the manufacturer).</li>
    <li>minage – the minimum recommended age to play.</li>
    <li>users_rated – the number of users who rated the game.</li>
    <li>average_rating – the average rating given to the game by users. (0-10)</li>
    <li>total_weights – Number of weights given by users. Weight is a subjective measure that is made up by BoardGameGeek. It’s how “deep” or involved a game is. http://boardgamegeek.com/wiki/page/Weight Here’s a full explanation</a>.</li>
    <li>average_weight – the average of all the subjective weights (0-5).</li>
</ul>


```python
board_data = pd.read_csv('data/games.csv')
board_data = board_data[board_data["average_rating"] > 0]
board_data.iloc[[5]]
```

## Discretization

Let's talk about some of the columns in this dataset. Most look like numeric variables but are they? See the min and max players columns. Perhaps this can be more thought of as a nominal variable, for example, 2-4 players, or 4-6 players. 

Feature engineering can be thought of as an art since there are a lot of intuitions a modeller can inject to his/her data science project. However, it is also a science since further along the pipeline, <b> these intuitions are put under the scrutiny of statistical metrics to determine if it is indeed useful or not </b>.

Let's plot the min players and max players' histogram. Then we combine the two to make a new column named "min-max players".

```python
plt.figure(figsize=(20,5))
c = Counter(board_data["minplayers"].dropna())
cats = c.keys()
freqs = [v for _,v in c.iteritems()]
plt.bar(range(len(c)), freqs)
plt.xticks(range(len(c)), cats)
plt.show()
```

```python
plt.figure(figsize=(20,5))
c = Counter(board_data["maxplayers"].dropna())
cats = c.keys()
freqs = [v for _,v in c.iteritems()]
plt.bar(range(len(c)), freqs)
plt.xticks(range(len(c)), cats)
plt.show()
```

```python
minplayers = board_data["minplayers"].tolist()
maxplayers = board_data["maxplayers"].tolist()
minmaxplayers = ["{}-{}".format(a,b) for (a,b) in zip(minplayers, maxplayers)]
```

```python
board_data['minmaxplayers'] = pd.Series(minmaxplayers, index=board_data.index)
print("Min max players data type: {}".format(board_data['minmaxplayers'].dtype))
board_data[:2][['name','minmaxplayers','minplayers','maxplayers']]
```

```python
plt.figure(figsize=(20,5))
c = Counter(board_data["minmaxplayers"].dropna())
cats = c.keys()
freqs = [v for _,v in c.iteritems()]
plt.bar(range(len(c)), freqs)
plt.xticks(range(len(c)), cats)
plt.show()
```

But still, this does not make sense. Let's group "discretize" min players, then max players, we will use our board game geekiness to understand that grouping together 1-2, 1-6, 1-10 and 1-above 10, etc. makes sense.

```python
min1players = board_data['minplayers'] == 1
min2players = board_data['minplayers'] == 2
min4players =(board_data['minplayers'] <=4) & (board_data['minplayers'] >2)
min_above4players =board_data['minplayers'] >4
```

```python
max2players = board_data['maxplayers'] <=2
max6players = (board_data['maxplayers'] <=6) & (board_data['maxplayers'] >2)
max_above6players = (board_data['maxplayers'] >6)
```

```python
board_data['Players'] = None
board_data.loc[min1players & max2players,'Players'] = '1-2 players'
board_data.loc[min1players & (max6players | max_above6players), 'Players'] = '1-above 2 players'
board_data.loc[min2players & (max2players | max6players), 'Players'] = '2-6 players'
board_data.loc[min2players & (max_above6players), 'Players'] = '2-above 6 players'
board_data.loc[min4players & max6players, 'Players'] = '4-6 players'
board_data.loc[min4players & max_above6players,'Players'] = '4-above 6 players'
board_data.loc[min_above4players & max6players, 'Players'] = 'above-4 to 6 players'
board_data.loc[min_above4players & max_above6players, 'Players'] = 'above-4 to above 6 players'
```

```python
plt.figure(figsize=(20,5))
c = Counter(board_data["Players"].dropna())
cats = sorted(c.keys())
freqs = [c[cat] for cat in cats]
plt.bar(range(len(c)), freqs, log=True)
plt.xticks(range(len(c)), cats)
plt.show()
```

Now that we have the histogram more understandable, let's now paint each bin if they are either "better" or not.

```python
players_average_rating = board_data.groupby('Players')['average_rating'].mean().to_dict()
board_data.groupby('Players', as_index=False)['average_rating'].mean()
```

```python
c = Counter(board_data["Players"].dropna())
cats = sorted(c.keys())
freqs = [c[cat] for cat in cats]

fig = plt.figure(figsize=(20,5))
fig.get_axes()
plt.bar(range(len(c)), freqs, log=True)
ax = fig.get_axes()[0]
ax2 = ax.twinx()

players_average_rating = board_data.groupby('Players')['average_rating'].mean().tolist()

ax2.plot(players_average_rating, c='r')
ax2.set_ylabel("Average rating")
plt.xticks(range(len(c)), cats)
ax.set_ylabel("Number of games")
plt.show()
```

Now we know that the bins above 4- above 6 players is the highest bin for the average rating. Not surprising since Resistance is in this bin. The importance of this step is that maybe this feature would be deemed by our learning algorithm to be a good indicator of ratings.

Finally, let's just run pandas to study every possible combination in our dataset.

```python
import operator
correlations= board_data.corr()["average_rating"].to_dict()
sorted(correlations.items(), key=operator.itemgetter(1), 
       reverse=True)
```

And what about the variable that we just did? Turns out, it became more useful when combined (minplayers and maxplayers are all 0.00's).

```python
from scipy.stats import pearsonr
pearsonr(pd.Categorical(board_data["Players"]).labels, board_data['average_rating'])
```

Let's now plot the best variables against average rating that we have.

To have our pretty color variable representing the average ratings, we should normalize it to the [0,1] range. I'll be using here the technique called min-max normalization.

```python
rating_series = board_data["average_rating"]
norm_ratings = (rating_series-min(rating_series))/(max(rating_series)-min(rating_series))
```

```python
from pandas.tools.plotting import scatter_matrix
scatter_matrix(board_data[["average_rating",
                           "average_weight",
                           "bayes_average_rating",
                           "minage",
                           "total_wanters"]], 
               alpha=0.2, figsize=(20, 20),
               diagonal='kde', marker='o', cmap='gray')
plt.show()
```

```python
fig = plt.figure(figsize=(20,15))
plt.scatter(board_data["average_weight"].tolist(), board_data["total_wanters"].tolist(), 
            s=200, marker='x', color=norm_ratings)
plt.colorbar()
plt.xlabel("Average weight")
plt.ylabel("Total Wanters")
plt.ylim((-200,1800))
plt.show()
```
